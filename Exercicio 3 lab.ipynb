## Exercício 1

```python
print("Qual a primeira nota?")
nota1 = float(input())

print("Qual a segunda nota?")
nota2 = float(input())

media = (nota1 + nota2) / 2

if 7 <= media < 10:
  print("Aprovado")

elif media < 7:
  print("Reprovado")

elif media == 10:
  print("Aprovado com distinção")
```

## Exercício 4

```python
numero = int(input("Digite o número de termos da sequência de Fibonacci que você deseja ver: "))

a, b = 1, 1

print("Série de Fibonacci:")
for i in range(numero):
    print(a, end = " ")
    a, b = b, a + b
```

## Exercício 5

```python
while True:
    nome = input("Digite seu nome: ")
    if len(nome) > 3:
        break
    print("Erro!")

while True:
    idade = int(input("Digite sua idade: "))
    if 0 < idade < 150:
        break
    print("Erro!")

while True:
    salario = float(input("Digite seu salário: "))
    if salario > 0:
        break
    print("Erro!")

while True:
    sexo = input("Digite seu sexo (digite f para feminino ou m para masculino): ").lower()
    if sexo in ('f', 'm'):
        break
    print("Erro!")

while True:
    estado_civil = input("Digite seu estado civil (digite s para solteiro, c para casado, v para viúvo(a) ou d para divorciado(a)): ").lower()
    if estado_civil in ('s', 'c', 'v', 'd'):
        break
    print("Erro!")

print("Informações validadas com sucesso!")
```

## Exercício 6

```python
numero = int(input("Digite um número inteiro: "))

if numero == 1:
    print(f"{numero} não é um número primo.")

elif numero == 2:
    print(f"{numero} é um número primo.")

else:
    for i in range(2, numero):
        if numero % i == 0:
            print(f"{numero} não é um número primo.")
            break
    else:
        print(f"{numero} é um número primo.")
```
